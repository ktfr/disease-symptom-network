```python
# imports
import pandas as pd
import os
import kagglehub
import numpy as np
# import kaggle 
```

```python
df_name = "dhivyeshrk/diseases-and-symptoms-dataset"
path = kagglehub.dataset_download(df_name)
```

```python
print("Dataset downloaded to:", path)
```

```python
file_path = os.path.join(path, 'Final_Augmented_dataset_Diseases_and_Symptoms.csv')  # Update with actual filename

# Load the dataset directly into a pandas DataFrame
print(f"Loading dataset from: {file_path}")
df = pd.read_csv(file_path)

# Display the first few rows of the dataset
print(df.head())
```

```python
df_aggregated = df.groupby("diseases").max().reset_index()
df_aggregated
```

```python
from itertools import combinations

disease_to_symptoms = df_aggregated.set_index("diseases")[df_aggregated.columns[2:]].to_dict(orient="index")

edges = []
for disease1, disease2 in combinations(disease_to_symptoms.keys(), 2):
    symptoms1 = np.array(list(disease_to_symptoms[disease1].values()))
    symptoms2 = np.array(list(disease_to_symptoms[disease2].values()))

    shared_symptoms = np.sum((symptoms1 == 1) & (symptoms2 == 1))

    if shared_symptoms > 0:
        edges.append((disease1, disease2, shared_symptoms))

# Convert to DataFrame and save outputs
edges_df = pd.DataFrame(edges, columns=["Disease1", "Disease2", "Weight"])
edges_df.to_csv("disease_network_edges.txt", sep="\t", index=False, header=False)
#df_aggregated.to_csv("diseases.csv", index=True)
edges_df.to_csv("disease_network_edges.csv", index=False)
```

```python
# grab diseases only
df_aggregated["diseases"].to_csv("diseases.csv", index=True)
```

```python
df_diseases_only = df_aggregated["diseases"]
```

```python
# Convert Series to DataFrame and create an explicit index
df_diseases_only = df_aggregated[["diseases"]].reset_index()

# Rename columns for clarity
df_diseases_only.columns = ["index", "disease_name"]

# Create the mapping dictionary
disease_name_to_index = df_diseases_only.set_index("disease_name")["index"].to_dict()

# Replace disease names with indices in edges_df
edges_df["Disease1"] = edges_df["Disease1"].map(disease_name_to_index)
edges_df["Disease2"] = edges_df["Disease2"].map(disease_name_to_index)

# Save the updated dataset
edges_df.to_csv("disease_network_edges_indexed.csv", index=False)
```
