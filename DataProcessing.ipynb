```python
# imports
import pandas as pd
import os
import kagglehub
import numpy as np
# import kaggle 
```

```python
file_path = os.path.join(path, 'Final_Augmented_dataset_Diseases_and_Symptoms.csv')  # Update with actual filename

# Load the dataset directly into a pandas DataFrame
print(f"Loading dataset from: {file_path}")
df = pd.read_csv(file_path)

# Display the first few rows of the dataset
print(df.head())
```

```python
df_aggregated = df.groupby("diseases").max().reset_index()
df_aggregated
```

```python
from itertools import combinations

edges = []

#gets unique diseases and then convert to python list
diseases = df_aggregated["diseases"].tolist()

symptom_columns = df_aggregated.columns[2:]

for disease1, disease2 in combinations(diseases, 2):
    #filter for the disease we are looking at, 
    #select only symptom related columns then convert the df row into an array
    symptoms1 = df_aggregated[df_aggregated["diseases"] == disease1][symptom_columns].values[0]
    symptoms2 = df_aggregated[df_aggregated["diseases"] == disease2][symptom_columns].values[0]

    shared_symptoms = np.sum((symptoms1 == 1) & (symptoms2 == 1))

    if shared_symptoms > 0:
        edges.append((disease1, disease2, shared_symptoms))

edges_df = pd.DataFrame(edges, columns=["Disease1", "Disease2", "Weight"])

edges_df.to_csv("disease_network_edges.txt", sep="\t", index=False, header=False)
```

```python
edges_df_sorted = edges_df.sort_values(by= "Weight", ascending = False)
edges_df_sorted
%store edges_df_sorted
%store edges_df
%store df_aggregated
```
