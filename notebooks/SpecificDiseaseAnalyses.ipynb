```python
# If there's a column like 'Disease' or 'Label', use that to filter
hypokalemia_symptoms = df[df['diseases'].str.lower() == 'hypokalemia'.lower()]

# First, filter the row(s) for 'hypokalemia'
hypokalemia_row = df[df['diseases'].str.lower() == 'hypokalemia']

# Then, drop non-symptom columns (like 'Label', 'Id', 'Community', etc.)
# This assumes that symptom columns are binary (0/1)
symptom_data = hypokalemia_row.drop(columns=['Label', 'Id', 'Community'], errors='ignore')

# Now, get the columns where the value is 1
symptoms_present = symptom_data.loc[:, (symptom_data == 1).any(axis=0)].columns.tolist()

# Print the symptom names
print("Symptoms for hypokalemia:")
for symptom in symptoms_present:
    print("-", symptom)

```
