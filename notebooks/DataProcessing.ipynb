```python
df_aggregated = df.groupby("diseases").max().reset_index()
df_aggregated
```

```python
from itertools import combinations
import numpy as np
import pandas as pd

edges = []

# Get unique diseases
diseases = df_aggregated["diseases"].tolist()
symptom_columns = df_aggregated.columns[2:]

for disease1, disease2 in combinations(diseases, 2):
    # Get symptom vectors as numpy arrays
    symptoms1 = df_aggregated[df_aggregated["diseases"] == disease1][symptom_columns].values[0]
    symptoms2 = df_aggregated[df_aggregated["diseases"] == disease2][symptom_columns].values[0]

    # cosine similarity
    dot_product = np.sum(symptoms1 * symptoms2)
    norm1 = np.sqrt(np.sum(symptoms1))
    norm2 = np.sqrt(np.sum(symptoms2))

    # so we don't divide the dot product by 0
    if norm1 == 0 or norm2 == 0:
        #if either of them are 0, then define the cosine similarity as 0 so we don't run into error
        cosine_sim = 0
    else:
        cosine_sim = dot_product / (norm1 * norm2)

    # if there is positive cosine similarity, then add that edge to the network
    if cosine_sim > 0:
        edges.append((disease1, disease2, cosine_sim))
        
        
edges_df = pd.DataFrame(edges, columns=["Disease1", "Disease2", "Weight"])

edges_df.to_csv("disease_network_edges.txt", sep="\t", index=False, header=False)

```

```python
edges_df_sorted = edges_df.sort_values(by= "Weight", ascending = False)
edges_df_sorted
%store edges_df_sorted
%store edges_df
%store df_aggregated
```
